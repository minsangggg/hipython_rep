# Heatmap

데이터를 요약(합계, 평균 등등)한 결과를 색깔로 시각화  
요약- 피봇 테이블(탐색) - 인사이트 시각화

```python
# 요약
pivot_df=tip_df.pivot_table('tip','day','size')
# 시각화
sns.heatmap(pivot_df, annot=True)
```

```python
pivot_df.fillna(0, inplace=True)
```

```python
sns.heatmap(pivot_df, annot=True)
```

```python
#plotly.imshow()
px.imshow(
    pivot_df, x=pivot_df.columns, y=pivot_df.index.astype('str'),
    width=400,height=400, text_auto='.2e'
)
```

## parallel_coordinates

평행좌표 플롯

```python
iris_df=px.data.iris()
iris_df.head()
```

```python
px.parallel_coordinates(
    iris_df,
    color='species_id',
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=2
)
```

```python
px.parallel_coordinates(
    tip_df,
    color='size',  # 각 선의 색깔을 테이블 size(인원 수)에 따라 다르게 설정
    color_continuous_scale=px.colors.diverging.Tealrose,  # 연속형 색상 팔레트 (Tealrose 계열)
    color_continuous_midpoint=2
)
```
