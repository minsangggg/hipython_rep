# 그래프 결과 해석 

적은 금액을 결제한 손님이, 적은 팁을 준 경우가 많습니다.

### 인사이트:
1. 가장 흔한 소비 패턴
대부분의 손님이 비싸지 않은 식사를 하고,
→ 소액의 팁을 준다는 뜻입니다.

2. 데이터의 밀집 구간
   데이터가 몰려 있다는 건 이 구간이 빈도상 가장 일반적인 사례라는 뜻입니다.  
분석적으로는 “평균적인 소비자 행동”이라 볼 수 있어요.

3. 왜도(Skewness) 또는 비선형성 힌트
점들이 왼쪽 하단에 몰려 있고, 오른쪽으로 갈수록 퍼지는 모양이라면,
전체 관계가 선형이 아닐 수도 있다는 신호입니다.
예를 들어: total_bill이 클 때 tip이 비례해서 오르지 않는다는 뜻일 수도 있어요.



```python
#boxplot
fig=plt.figure(figsize=(12,5))
axes=fig.add_subplot(1,1,1) #totall_bill
axes.boxplot(data=tips_df,x='total_bill')
plt.show
```

```python
fig=plt.figure(figsize=(12,5))
axes=fig.add_subplot(1,1,1) #totall_bill
axes.boxplot(
    x=[
      tips_df[tips_df['sex']=='Female']['total_bill'],
      tips_df[tips_df['sex']=='Male']['total_bill']    #지불하는 폭이 더 넓다, 중앙값 다르고, 최대값이 남성이 더 크고 
    ],
    labels=['Female','Male']
)
plt.show()
```

```python
#  남성부분에서 중앙값이 박스 중앙에 있지 않고 한쪽으로 치우쳐 있으므로 비대칭 분포이다.
# 전체 데이터에서 가장 큰 값은 이상치에서 찾아야한다
```

```python
Female_total_bill = tips_df[tips_df['sex'] == 'Female']['total_bill']

# 2. 사분위수 계산
Q1 = tips_df[tips_df['sex']=='Female']['total_bill'].quantile(0.25)
Q3 = tips_df[tips_df['sex']=='Female']['total_bill'].quantile(0.75)
IQR = Q3 - Q1

# 3. 이상치 기준 계산
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# 4. 이상치 조건
cond1 = tips_df[tips_df['sex']=='Female']['total_bill'] < lower
cond2 = tips_df[tips_df['sex']=='Female']['total_bill'] > upper

# 5. 이상치 개수 출력
print("이상치 개수:", (cond1.sum() + cond2.sum()))
```

```python
colors={"Female":"red", "Male":"blue"}

#다변량 표현
#성별에 따라-색상
#값의 크기에 따라-점의 크기
fig=plt.figure(figsize=(12,5))
axes=fig.add_subplot(1,1,1) 
axes.scatter(
  tips_df['total_bill'], tips_df['tip'],        
  s=tips_df['size']**2*10,
  c=tips_df['sex'].map(colors)   
     )
plt.show()   
```

```python
#colors={"Female":"red", "Male":"blue"}
colors={"Female":"#ff00ad","Male":"#000080"}
tips_df['tip_pct']=tips_df['tip']/tips_df['total_bill']

fig=plt.figure()
axes=fig.add_subplot(1,1,1) 
axes.scatter(
  tips_df['total_bill'], tips_df['tip'],          #순서대로 x축,y축
  s=tips_df['tip_pct']*1000,                        #점의 크기, 식사인원수(size)에 제곱해서 10배
  c=tips_df['sex'].map(colors),                    #남자는 파랑, 여자는 빨강
     alpha=0.5
     )
plt.show()   
```

```python
#파이 차트:비율 파악
#성별 비율 비교
#빈도수

sex_counts=tips_df['sex'].value_counts()
sex_counts.head()
fig=plt.figure()
axes=fig.add_subplot(1,1,1) 
axes.pie(
  sex_counts,
  colors=['blue','red'],
  labels=sex_counts.index,
  autopct='%1.1f%%',
  startangle=90
     )
plt.show()   
```

```python
size_counts=tips_df['size'].value_counts()
size_counts.head()
fig=plt.figure()
axes=fig.add_subplot(1,1,1) 
axes.pie(
  size_counts,
  colors=['red','orange','yellow','green','blue','purple'],
  labels=size_counts.index,
  autopct='%1.1f%%',
  startangle=100
     )
plt.show()   
```
