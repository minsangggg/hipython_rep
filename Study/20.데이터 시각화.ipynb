```python
import matplotlib.pyplot as plt
```

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
plt
```

```python
plt.figure(figsize=(10,6))        #가로,세로 길이
plt.title('5일간 주식 가격')
plt.plot(dates,prices,marker='o',label='주식 가격',color='blue')

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()                        #격자
plt.grid()
plt.ylim(95,110)
plt.xlim(0,4)
```

```python
plt.figure(figsize=(10,6))        #가로,세로 길이
plt.title('5일간 주식 가격')
plt.ylim(95,110)
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()                         #격자


plt.plot(dates,prices, label='Stock Price??')
plt.legend()
plt.show()

```

```python
plt.subplot(1,2, 1)
plt.plot(dates, prices) #line 
plt.xticks(rotation=40)

plt.subplot(1,2, 2)
plt.bar(dates, prices)
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart.png')

#plt.subplot(1,2,1)
#plt.plot(dates, prices)
#plt.plot(dates,prices)

#plt.xticks(rotation=10)

#plt.subplot(1,2,2)
#plt.plot(dates, prices)
#plt.plot(dates,prices)


#plt.savefig('./data/')
```

```python
plt.subplot(1,2,1)
plt.plot(dates,prices, marker='o',color='green')
plt.title("Stock Prices-Subplot 1")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.bar(dates,prices, color='orange')
plt.title("Stock Prices-Subplot 2")
plt.xticks(rotation=45)

plt.show()
plt.savefig("stock_prices_5_dats.png")
```
