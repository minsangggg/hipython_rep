## stratified split

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns=iris.feature_names )
iris_df['label']  = iris.target # 각 행(샘플)이 어떤 품종(클랫, 정답)에 해당하는지 label이라는 새컬럼에 추가
iris_df['label'].value_counts() #label 컬럼에 각 클래스(0,1,2)가 몇 개씩 있는지
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris_df, iris_df['label']):
  y_train = iris_df['label'].iloc[train_index]
  y_test =  iris_df['label'].iloc[test_index]
  print(f'train 분포: {y_train.value_counts()}')
  print(f'test 분포: {y_test.value_counts()}')
```

## GridSearchCV- 하이퍼파라미터+교차검증

```python
x_train, x_test, y_train,y_test=train_test_split(iris_data,iris_label,
                                                 test_size=0.2,
                                                 random_state=121)
```

```python
from sklearn.model_selection import GridSearchCV
dt_clf2=DecisionTreeClassifier(random_state=121) #의사결정나무 분류모델 새로 만듦

#max_depth:트리 최대 깊이, min_samples_split: 노드를 분할하는데 걸린 시간
params= {'max_depth':[1,2,3], 'min_samples_split':[2,3]}

#의사결정나무 모델을 다양한 파라미터 조합으로 3번씩 교차검증해
#가장 좋은 조합을 자동으로 찾아주는 도구(GridSearchCV) 객체를 만든다.
grid_dtree=GridSearchCV(dt_clf2,param_grid=params, cv=3, refit=True)

#모든 파라미터 조합에 대해 모델을 학습하고, 교차검증으로 최적조합 자동선택
grid_dtree.fit(x_train,y_train) 
```

```python
pd.DataFrame(grid_dtree.cv_results_)
```

```python
b_model=grid_dtree.best_estimator_
pred=b_model.predict(x_test)
accuracy_score(y_test,pred)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# titanic.csv 파일 읽어오기 (파일 경로는 필요에 따라 조정)
df = pd.read_csv('./data/titanic.csv')
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})  # ★ 여기서 변환!
df['Age'] = df['Age'].fillna(df['Age'].median())

# 결측치 먼저 채우고
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 원핫 인코딩 (S, C, Q를 각각 숫자 컬럼으로 변환)
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
x=df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked_C','Embarked_Q','Embarked_S']]
y=df['Survived']

#훈련데이터, 테스트데이터 분리
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.2,
                                               random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
dt_clf2=DecisionTreeClassifier(random_state=121) #의사결정나무 분류모델 새로 만듦

#max_depth:트리 최대 깊이, min_samples_split: 노드를 분할하는데 걸린 시간
params= {'max_depth':[1,2,3,4,5,6,7], 'min_samples_split':[2,3,4]}

#의사결정나무 모델을 다양한 파라미터 조합으로 3번씩 교차검증해
#가장 좋은 조합을 자동으로 찾아주는 도구(GridSearchCV) 객체를 만든다.
grid_dtree=GridSearchCV(dt_clf2,param_grid=params, cv=3, refit=True)

#모든 파라미터 조합에 대해 모델을 학습하고, 교차검증으로 최적조합 자동선택
grid_dtree.fit(x_train,y_train) 
```

```python
b_model=grid_dtree.best_estimator_
pred=b_model.predict(x_test)
accuracy_score(y_test,pred)
```
