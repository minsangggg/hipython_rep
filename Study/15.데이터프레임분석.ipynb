```python
#기대수명 평균이 가장 낮은 국가는? 대륙은?
#df.groupby(['continent'])[['lifeExp']].mean()
#df.groupby(['continent'])[['lifeExp']].mean().idxmin()

#대륙
df.groupby(['country'])[['lifeExp']].mean().idxmin()       #idxmin을 통해 인덱스로 접근하여 가장낮은 값을 알려준다
                                                           #groupby뒤에 ( ) 소괄호를 먼저쓰는 이유는 함수이기 때문이다
```

```python
g_df_mean=df.groupby(['year'])[['lifeExp']].mean()
g_df_mean.plot()                 #시간이 지날수록 기대수명이 올라간다 
```
